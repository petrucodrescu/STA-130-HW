# Question 5

The difference between an attribute and a method comes to what their purpose is and how they are called. The first distinction is an attribute does not need parentheses to be called, an example is .shape. While a method does need parantheses such as .describe(). The second distinction is what they return or their purpose. For the attribute it **provides** information about the data frame such as the number of columns or how many people values are in a row. A method usually has to **computate** values from the data. For example .describe() calculates the mean when called or the quartiles of the numerical data. A side note as well, when methods have empty parantheses, it usually means it has a default argument to work with. Once again for example, .describe()'s default value is all the columns with numerical values.
